```python
# DATA LINK: ADZUNA API
# I am using the ADZUNA API to get job data about the salaries, and number of applications
# I store it in a json and return it.

import requests
from bs4 import BeautifulSoup
import pandas as pd

url = f'https://api.adzuna.com/v1/api/jobs/us/search/1'
api_key = 'd429c47ed52836026f8e05afefb7fd07'
app_id = '1a701462'
locations = ['Boston, MA', 'Seattle, WA', 'New York City, NY', 'Austin, TX', 'San Diego, CA', 'Palo Alto, CA', 'Denver, CO', 
            'San Jose, CA', 'Washington, D.C.', 'Dallas, TX', 'Los Angelers, CA', 'San Francisco, CA', 'Chicago, IL', 'Atlanta, GA', 
            'Portland, OR', 'Raleigh, NC', 'Phoenix, AZ', 'Salt Lake City, UT', 'Minneapolis, MN', 'Pittsburgh, PA', 'Philadelphia, PA', 
            'Miami, FL', 'Indianapolis, IN', 'Charlotte, NC', 'Nashville, TN', 'Madison, WI', 'Buffalo, NY', 'Rochester, NY', 'Detroit, MI' ]


def get_data(job_position):
    data = pd.DataFrame()
    """ Get the data for all the location

    Args job_position (String): The job name

    Returns: data (DataFrame): a dataframe of all the data of the jobs

    """

    for location in locations: 
        def scrape_data(job_position, location):
            """ Get the data for one location

            Args: 
                job_postion (String): The job i want to get data for
                location (String): The location of the data.
            
            Returns:
                df (DataFrame): Data frame of the data
            """

            # Set params
            params = {
                'app_id': app_id,
                'app_key': api_key,
                'results_per_page': 100,
                'what': job_position, 
                'where': location,
                'content-type': 'application/json'
            }

            # Get data and create param series
            response = requests.get(url=url, params=params)
            job_data = response.json()['results']
            job_name = pd.Series()
            loc = pd.Series() 
            salary = pd.Series()
            applications = pd.Series()

            # Get data and add it to the series's
            for idx, job in enumerate(job_data):
                job_name.at[idx] = job.get('title', 'N/A')
                loc.at[idx] = location
                salary.at[idx] = job.get('salary', 'Not listed')
                applications.at[idx] = job.get('applications', 'Not available')
            
            # Add to the data frame
            df = pd.DataFrame({
                'Job Name':job_name,
                'Location':loc, 
                'Salary':salary,
                'Applications':applications})
            
            return df
        data = pd.concat([data, scrape_data('Software Engineer', location)], ignore_index=True)
    return data

get_data('Software Engineer').to_json('data.json', index=False) 
```
